```python
#import libraries
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
import sys
import docx2txt
import matplotlib.pyplot as plt
%matplotlib inline
import random
#setting maximum row of 1000 rows as result display
pd.set_option('display.max_row', 1000)

# Common imports
import numpy as np
import os
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
# to make this notebook's output stable across runs
np.random.seed(42)
```

```python
result = docx2txt.process("reviews.docx")
```

```python
type(result)
```

```python
def preprocess(result):
    result = result.replace("\n\n", " ")
    result = result.replace("(<br/>)", "")
    result = result.replace('(<a).*(>).*(</a>)', '')
    result = result.replace('(&amp)', '')
    result = result.replace('(&gt)', '')
    result = result.replace('(&lt)', '')
    result = result.replace('(\xa0)', ' ')  
    return result
```

```python
reviews = preprocess(result)
```

```python
reviews
```

```python
foo = list(reviews.split(" "))
```

```python
random.shuffle(foo)
```

```python
foo
```

```python
shfl_reviews = ' '.join(foo)
```

```python
shfl_reviews
```

```python
shfl_reviews = shfl_reviews.replace("  ", " ")
```

```python
shfl_reviews
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(shfl_reviews)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("2.png")
```
